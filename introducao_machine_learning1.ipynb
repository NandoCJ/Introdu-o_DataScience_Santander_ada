```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

!pip install seaborn
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
x_df= df[['total_bill']]  ##caracteristicas
y_df= df['tip'] ##alvo
```

```python
x_treino, x_teste,y_treino, y_teste = train_test_split(x_df,y_df)
```

```python
modelo = LinearRegression()
```

```python
modelo_treino=modelo.fit(x_treino,y_treino)##treinamento/ajuste da formula usando os dados do treino
```

```python
y_previsto = modelo.predict(x_teste)
```

```python
y_previsto
```

```python
df['tip'].mean()
```

```python
mean_absolute_error(y_previsto,y_teste)
```
